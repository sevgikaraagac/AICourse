```python
import numpy as np
import matplotlib.pyplot as plt
def generate_data(n, beta_0, beta_1):
    x = np.arange(n)
    e = np.random.uniform(-10,10, size=(n,))
    y = beta_0 + beta_1* x + e
    return x,y

x, y = generate_data(100, 2, .4)
plt.figure(figsize=(8,6))
plt.title("Input Data for Regression")
plt.scatter(x,y)

```

```python

#y=mx+b  
m=0;
b=0;
Learning_Rate= 0.0001
iteration=1000

n=100

for i in range(iteration):
    y_pred=m*x+b
    mse=(1/n)* np.sum((y-y_pred)**2)
    derrivative_m=(-2/n)*np.sum(x*(y-y_pred))
    derrivative_b=(-2/n)*np.sum(y-y_pred)
    m=m-Learning_Rate * derrivative_m
    b=b-Learning_Rate * derrivative_b
       
x_line=np.linspace(0, 100, 100)
y_line=m*x_line+b
plt.figure(figsize=(8,6))
plt.title("Regression Line")
plt.scatter(x,y)
plt.plot(x_line,y_line )
print("Mean Squared Error:")
print(mse)
plt.show()
```
